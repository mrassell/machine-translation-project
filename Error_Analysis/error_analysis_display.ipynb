```python
import pandas as pd

# =========================
# 1. Loading in the dataset
# =========================

fPath = r"error_analysis_dev_seed6761.csv"
df = pd.read_csv(fPath)

print("\nSupplied dataset:")
print(df.shape)
print(df.columns)
```

```python
# =========================
# 2. Cleaning the set
# =========================

df["category"] = df["category"].astype(str)
df["severity"] = df["severity"].astype(str)
df["system"] = df["system"].astype(str)
```

```python
# =========================
# 3. Displaying error distribution
# =========================

eCounts = df["category"].value_counts().reset_index()
eCounts.columns = ["Error Category", "Count"]
eCounts["Percentage"] = (
    eCounts["Count"] / eCounts["Count"].sum() * 100
)

print("\n=== Error Distribution ===")
print(eCounts)


# Save for report
eCounts.to_csv("error_distribution.csv", index=False)
```

```python
# =========================
# 4. Severity breakdown
# =========================

sevTable = pd.crosstab(df["category"], df["severity"])

print("\n=== Severity Breakdown ===")
print(sevTable)

sevTable.to_csv("severity_breakdown.csv")
```

```python
# =========================
# 5. Comparing systems
# =========================

sysTable = pd.crosstab(df["system"], df["category"])

print("\n=== System vs Error Category ===")
print(sysTable)

sysTable.to_csv("system_error_comparison.csv")
```

```python
# =========================
# 6. Metric analysis by error type
# =========================

mSum = df.groupby("category")[["bleu", "chrf", "comet"]].mean()

print("\n=== Metric Averages by Error Category ===")
print(mSum)

mSum.to_csv("metric_by_error_type.csv")
```

```python
# =========================
# 7. Severity and metrics
# =========================

sevMetrics = df.groupby("severity")[["bleu", "chrf", "comet"]].mean()

print("\n=== Metrics by Severity ===")
print(sevMetrics)

sevMetrics.to_csv("metric_by_severity.csv")
```

```python
# =========================
# 8. Top error subcategories
# =========================

if "category" in df.columns:
    eTop = df["category"].value_counts().head(15)

    print("\n=== Top Error Types ===")
    print(eTop)

    eTop.to_csv("top_errors.csv")
```

```python
# =========================
# 9. Qualitative sampling log
# =========================

def sample_errors(category, severity=None, n=5):
    subset = df[df["category"] == category]

    if severity:
        subset = subset[subset["severity"] == severity]

    return subset.sample(min(n, len(subset)))


print("\n=== Sample: Accuracy Major Errors ===")
print(sample_errors("Accuracy/Addition", "Major", 5)[
    ["source", "mt", "reference", "system"]
])
```

```python
# =========================
# 10. Formatted summary
# =========================

with open("mt_error_summary.md", "w", encoding="utf-8") as f:
    f.write("# MT Error Analysis Summary\n\n")

    f.write("## Error Distribution\n")
    f.write(eCounts.to_markdown(index=False))
    f.write("\n\n")

    f.write("## Severity Breakdown\n")
    f.write(sevTable.to_markdown())
    f.write("\n\n")

    f.write("## System Comparison\n")
    f.write(sysTable.to_markdown())
    f.write("\n\n")

    f.write("## Metric Summary by Error Type\n")
    f.write(mSum.to_markdown())
    f.write("\n")


print("\nAnalysis complete. Files exported.")
```
