```python
import pandas as pd
df = pd.read_csv('mqm_generalMT2023_ende.tsv', sep='\t')
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
ENTITY_TYPES = {"PERSON", "ORG", "GPE"}
NUMERIC_ENTITY_TYPES = {"DATE", "TIME", "MONEY", "PERCENT", "QUANTITY", "CARDINAL"}

def compute_entity_counts(texts, nlp, entity_types, numeric_entity_types):
    entity_counts = []
    numeric_counts = []

    for doc in nlp.pipe(texts, batch_size=64):
        ent_count = 0
        num_count = 0

        for ent in doc.ents:
            if ent.label_ in entity_types:
                ent_count += 1
            if ent.label_ in numeric_entity_types:
                num_count += 1

        entity_counts.append(ent_count)
        numeric_counts.append(num_count)

    return entity_counts, numeric_counts

```

```python
df["entity_count"], df["numerical_entity_count"] = compute_entity_counts(
    df["source"].tolist(),
    nlp,
    ENTITY_TYPES,
    NUMERIC_ENTITY_TYPES
)

df["stress_score"] = df["entity_count"] + df["numerical_entity_count"]
df.to_csv("wmt_mqm_2023_scored.csv", index=False)
```
